## Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Learn Excel\Tableau Datasets\Web Scrapping using Python and Pandas\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('classic')
```

```python
df.plot(kind='line', title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Scores')
```

```python
df['Flavor Rating'].plot(kind='bar')
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.scatter(x = 'Texture Rating', y= 'Overall Rating', s= 120, c = 'Orange')
```

```python
df.plot.hist()
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (20, 10))
```
